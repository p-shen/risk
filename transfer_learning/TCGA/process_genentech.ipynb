```r
options(stringsAsFactors = F)

tpm <- read.table("../TCGA//data/Genetech_expression_TPM.txt", sep="\t")
tpm <- t(tpm)
tpm <- as.data.frame(tpm)
```

```r
pd_genes <- c("CD274", "PDCD1")
```

```r
genes.tr <- read.table("/home/pzs2/capstone/proj/TCGA_processed/pancancer_all_immune/geneset.txt", header = T)[,1]
genes <- c(genes.tr,pd_genes)
tpm <- tpm[,(colnames(tpm) %in% genes.tr)]
```

```r
genes[which(!genes %in% colnames(tpm))]
```

```r
tpm$NSG1 <- 0.01
tpm <- tpm[,genes.tr]
```

```r
clin <- read.table("data/IMvigor210.clinical", sep="\t")
```

```r
merged <- merge(tpm, clin, by='row.names')
merged <- merged[,(colnames(merged) %in% c(genes.tr, 'OS','Event', 'Response'))]
```

```r
library(dplyr)
merged <- merged %>% filter(!is.na(Event)) %>% filter(!is.na(Response))
```

```r
sum(is.na(merged))
```

```r
genes.tr.data <- merged[,genes.tr]
labels.surv <- merged[,c('OS', 'Event')]
labels.resp <- merged$Response
```

```r
write.table(genes.tr.data, "data/genentech_processed.tsv", sep="\t", quote = F)
write.table(labels.surv, "data/genentech_processed.surv.tsv", sep="\t", quote = F)
write.table(labels.resp, "data/genentech_processed.labels.tsv", sep="\t", quote = F)
```
